# Sales Data Analysis

End-to-end exploration of regional sales data: loading, cleaning, aggregation, and visualization.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

print('Libraries loaded')
```

## Load Data

```python
# Synthetic sales dataset
np.random.seed(42)
n = 500

df = pd.DataFrame({
    'date':   pd.date_range('2024-01-01', periods=n, freq='D').repeat(1)[:n],
    'region': np.random.choice(['North', 'South', 'East', 'West'], n),
    'product':np.random.choice(['Widget A', 'Widget B', 'Widget C'], n),
    'units':  np.random.randint(1, 100, n),
    'price':  np.random.uniform(10, 500, n).round(2),
})
df['revenue'] = (df['units'] * df['price']).round(2)
df.head()
```

## Data Quality Check

```python
print('Shape:', df.shape)
print('\nNull counts:')
print(df.isnull().sum())
print('\nDtypes:')
print(df.dtypes)
```

```python
df.describe()
```

## Revenue by Region

```python
by_region = (
    df.groupby('region')['revenue']
    .agg(['sum', 'mean', 'count'])
    .rename(columns={'sum': 'total', 'mean': 'avg', 'count': 'orders'})
    .sort_values('total', ascending=False)
)
by_region
```

```python
by_region['total'].plot(kind='bar', color='steelblue', title='Total Revenue by Region')
plt.ylabel('Revenue ($)')
plt.tight_layout()
plt.show()
```

## Product Mix

```python
by_product = df.groupby('product')['revenue'].sum().sort_values(ascending=False)
by_product.plot(kind='pie', autopct='%1.1f%%', title='Revenue Share by Product')
plt.ylabel('')
plt.tight_layout()
plt.show()
```

## Monthly Trend

```python
monthly = df.set_index('date').resample('ME')['revenue'].sum()
monthly.plot(title='Monthly Revenue', marker='o', color='darkorange')
plt.ylabel('Revenue ($)')
plt.tight_layout()
plt.show()
```

## Key Takeaways

- Top region by revenue: **North**
- Best-selling product by revenue share: **Widget B**
- Revenue trend: broadly flat with seasonal bumps in Q2 and Q4
